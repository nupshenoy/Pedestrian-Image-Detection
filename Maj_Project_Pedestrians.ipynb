```python
import cv2
import sys
```

```python
pedsCascade=cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_fullbody.xml')
```

```python
imgloc = "ped1.jpg"
img = cv2.imread(imgloc)
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show()
```

```python
peds = pedsCascade.detectMultiScale(
            gray,
            scaleFactor = 1.003,
            minNeighbors = 2,
            minSize = (30,30)

)
```

```python
print("Found {0} pedestrians".format(len(peds)))
```

```python
for (x,y,w,h) in peds:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0,255,0), 2)
```

```python
status = cv2.imwrite('peds_found.jpg', img)
print("Image stored in system: ", status)
```
